```python
import matplotlib.pyplot as plt

x = np.array ([1,2,3,4,5])
y = np.array ([2,4,6,8,10])

plt.plot(x,y)
```

```python
plt.plot (x,y, marker = '*')
```

```python
    plt.plot (x,y, marker = '*', c = 'red')
```

```python
plt.plot (x,y, marker = '*', c = 'red', ds = 'steps', ls = '--')
```

```python
plt.plot (x,y, marker = '*', c = 'red', ds = 'steps', ls = '--', mec = 'blue')
```

```python
plt.plot(x,y)
plt.xlabel('Valores de X', fontsize = 12)
plt.ylabel('Valores de Y', fontsize = 12)
plt.title ('Título do gráfico', fontsize = 15, c='purple')
plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'orange')
```

```python
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)

# dois gráficos

plt.plot (x,y)
plt.plot (x,z)
```

```python
plt.plot(x,y, label = 'seno')
plt.plot (x,z,label='cosseno')
plt.plot (x,np.sqrt(x), label='raiz quadrada')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('subplots')
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```
